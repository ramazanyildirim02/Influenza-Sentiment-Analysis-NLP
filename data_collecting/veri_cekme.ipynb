```python
from docx import Document
import re

def cumleleri_ayir_ve_diziye_ekle(docx_path):
    doc = Document(docx_path)

    cumleler_dizisi = []
    for paragraph in doc.paragraphs:
        # Cümle ayırma için sadece noktaları kullanabilirsiniz
        cumle_listesi = re.split(r'\.\s', paragraph.text)
        
        # Eğer cumle_listesi boş değilse, cümleleri dizisine ekleyelim
        if cumle_listesi:
            cumleler_dizisi.extend([cumle.strip() for cumle in cumle_listesi])

    return cumleler_dizisi

# DOCX dosya yolu
dosya_yolu = "Grip.docx"

# Fonksiyonu çağırarak cümleleri alalım
cumleler_dizisi = cumleleri_ayir_ve_diziye_ekle(dosya_yolu)

# Cümleleri ekrana yazdıralım
for index, cumle in enumerate(cumleler_dizisi, start=1):
    print(f"Cümle {index}: {cumle}")
```
